```python
df_amazon.dropna(subset=['Time_Orderd'],inplace=True)
df_amazon.isna().sum()
```

```python

df_amazon['Time_Orderd'] = pd.to_timedelta(df_amazon['Time_Orderd'] + ':00')  # Add seconds for consistency
df_amazon['Time_Order_picked'] = pd.to_timedelta(df_amazon['Time_Order_picked'] + ':00')

# Adjust for cases where the picked time is earlier (next day deliveries)
df_amazon.loc[df_amazon['Time_Order_picked'] < df_amazon['Time_Orderd'], 'Time_Order_picked'] += pd.Timedelta(days=1)

# Calculate delivery time in minutes
df_amazon['Delivery_time'] = (df_amazon['Time_Order_picked'] - df_amazon['Time_Orderd']).dt.total_seconds() / 60
df_amazon['Time_Orderd'] = df_amazon['Time_Orderd'].apply(lambda x: str(x)[7:])  # Extracts HH:MM:SS
df_amazon['Time_Order_picked'] = df_amazon['Time_Order_picked'].apply(lambda x: str(x)[7:])  # Extracts HH:MM:SS
```

```python
# Remove NaN values
df_amazon = df_amazon.dropna(subset=["Delivery_time", "Weather", "Road_traffic_density"])

# Set figure size
plt.figure(figsize=(15, 5))

# 📊 1. Boxplot: Delivery Time vs Weather Conditions
plt.subplot(1, 3, 1)
sns.boxplot(x="Weather", y="Delivery_time", data=df_amazon, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Delivery Time vs Weather Conditions")

# 📊 2. Barplot: Average Delivery Time vs Traffic Conditions
plt.subplot(1, 3, 2)
sns.barplot(x="Road_traffic_density", y="Delivery_time", data=df_amazon, estimator=lambda x: sum(x) / len(x), palette="magma")
plt.title("Average Delivery Time vs Traffic Conditions")

# 📊 3. Scatterplot: Traffic Density & Weather vs Delivery Time
plt.subplot(1, 3, 3)
sns.scatterplot(x=df_amazon["Road_traffic_density"], y=df_amazon["Delivery_time"], hue=df_amazon["Weather"], palette="viridis")
plt.title("Traffic, Weather & Delivery Time")

plt.tight_layout()
plt.show()
```
