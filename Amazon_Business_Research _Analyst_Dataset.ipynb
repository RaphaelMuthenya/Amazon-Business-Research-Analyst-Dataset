 Rows with missing 'Time_Orderd' values were permanently removed from df_amazon.
 The remaining dataset was checked for any other missing values using isna().sum().
 This step ensures that the dataset is cleaned from critical missing values, improving accuracy in further analysis. 

```python
df_amazon.dropna(subset=['Time_Orderd'],inplace=True)
df_amazon.isna().sum()
```

```python
# Convert 'Time_Orderd' and 'Time_Order_picked' columns to timedelta format   
df_amazon['Time_Orderd'] = pd.to_timedelta(df_amazon['Time_Orderd'] + ':00')  
df_amazon['Time_Order_picked'] = pd.to_timedelta(df_amazon['Time_Order_picked'] + ':00')  

# Adjust for cases of next-day deliveries. If 'Time_Order_picked' is earlier than 'Time_Orderd', add 1 day to maintain correct time difference  
df_amazon.loc[df_amazon['Time_Order_picked'] < df_amazon['Time_Orderd'], 'Time_Order_picked'] += pd.Timedelta(days=1)  

# Calculate the delivery time in minutes  

df_amazon['Delivery_time'] = (df_amazon['Time_Order_picked'] - df_amazon['Time_Orderd']).dt.total_seconds() / 60  

# Convert timedelta back to string format 
df_amazon['Time_Orderd'] = df_amazon['Time_Orderd'].apply(lambda x: str(x)[7:])  
df_amazon['Time_Order_picked'] = df_amazon['Time_Order_picked'].apply(lambda x: str(x)[7:])  
```

## Exploratory Data Analysis and Visualization
This segment aims to analyze the relationship between delivery time and external factors such as weather conditions and road traffic density using data visualization techniques. The goal is to identify trends and patterns that could impact delivery efficiency.



```python
# Remove NaN values in the columns "delivery time","Weather" and "Road_traffic_density"
df_amazon = df_amazon.dropna(subset=["Delivery_time", "Weather", "Road_traffic_density"])

plt.figure(figsize=(20, 10))

# Box plot
plt.subplot(1, 3, 1)
sns.boxplot(x="Weather", y="Delivery_time", data=df_amazon, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Delivery Time vs Weather Conditions")

# Bar plot
plt.subplot(1, 3, 2)
sns.barplot(x="Road_traffic_density", y="Delivery_time", data=df_amazon, estimator=lambda x: sum(x) / len(x), palette="magma")
plt.title("Average Delivery Time vs Traffic Conditions")

# Scatter Plot
plt.subplot(1, 3, 3)
sns.scatterplot(x=df_amazon["Road_traffic_density"], y=df_amazon["Delivery_time"], hue=df_amazon["Weather"], palette="viridis")
plt.title("Traffic, Weather & Delivery Time")

plt.tight_layout()
plt.show()
```
