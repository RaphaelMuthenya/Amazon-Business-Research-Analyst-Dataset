```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
import seaborn as sns
df_amazon_file=pd.read_csv("cleaned_test.csv")
#viewing of all columns
df_amazon_file.columns
```

```python
df_amazon_file.head()
```

```python
#Data Cleaning: df_amazon_fileremove null columns but do not drop for use of future data
df_amazon_file = df_amazon_file.dropna(subset=["Time_Orderd", "Weather", "Road_traffic_density","Festival"])
```

```python
df_amazon_file.head()
```

```python
#check the columns in the dataset
df_amazon_file.columns
```

```python
#get totalcount of those delivered during festival and those delivered during non-festival
def totalCount():
    count_yes = (df_amazon_file['Festival'] == 'Yes').sum()
    count_No = (df_amazon_file['Festival'] == 'No').sum()

    return print("Those delivered during Festival Time:",count_yes,"\nThose delivered during Non-Festival Time:",count_No)

totalCount()

```

```python
#Count deliveries based on location that is the city column
deliveredcity_city = df_amazon_file["City"].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=deliveredcity_city.index, y=deliveredcity_city.values, palette="pastel")
plt.xticks(rotation=45)
plt.xlabel("Cities")
plt.ylabel("Total Deliveries")
plt.title("Busiest delivery locations")
plt.show()
```

```python
# Compare delivery performance on festival and non festival days
festival_delivery_counts = df_amazon_file["Festival"].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=festival_delivery_counts.index, y=festival_delivery_counts.values, palette="deep")
plt.xticks(rotation=45)
plt.xlabel("Festival Status")
plt.ylabel("Total Delivery")
plt.legend(title='Festival')
plt.title("Delivery Performance on festival and non festival days")
plt.show()
```

```python
#Pivot table to identify each city delivery count based on ordered date and festival season
pivot_table_df = df_amazon_file.pivot_table(index='City', columns='Festival', values='Road_traffic_density', aggfunc='count', fill_value=0)

print(pivot_table_df)
```
